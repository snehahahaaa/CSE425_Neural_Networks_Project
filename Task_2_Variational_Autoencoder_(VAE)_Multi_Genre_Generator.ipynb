```python
# CSE425 Task 2 - LSTM Variational Autoencoder for Multi-Genre MIDI Generation
# Extending Task 1 by replacing the plain autoencoder with a VAE so that the
# latent space becomes continuous and sampable from a standard normal prior.

import random
import shutil
import warnings
import hashlib
import zipfile
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pretty_midi  # handling MIDI file I/O

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader

from tqdm.auto import tqdm
from IPython.display import display

warnings.filterwarnings("ignore")

# Fixing all random seeds so results are reproducible across runs
seed = 42
random.seed(seed)
np.random.seed(seed)
torch.manual_seed(seed)

if torch.cuda.is_available():
    torch.cuda.manual_seed_all(seed)

# Selecting GPU if available, otherwise falling back to CPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Folder Setup
# Reading MIDI from Google Drive and writing all outputs to local Colab storage.
# Also pointing at the Task 1 ZIP so we can pull its metrics for comparison later.

drive_midi_folder = Path("/content/drive/MyDrive/CSE425_Project/task2_midi/multigenre")
task1_zip_path = Path("/content/drive/MyDrive/CSE425_Project/task1_outputs/CSE425_Task1_LSTM_Autoencoder.zip")

project_folder = Path("/content/CSE425_Task2_VAE_Music_Generator")

# Wiping any previous run so we always start clean
if project_folder.exists():
    shutil.rmtree(project_folder)

# Defining all output subdirectory paths
raw_folder = project_folder / "data" / "raw_midi"
processed_folder = project_folder / "data" / "processed"
output_folder = project_folder / "outputs"
midi_output_folder = output_folder / "generated_midis"
plot_folder = output_folder / "plots"
result_folder = output_folder / "results"
model_folder = output_folder / "model"

# Creating all folders in one pass
for folder in [raw_folder, processed_folder, midi_output_folder, plot_folder, result_folder, model_folder]:
    folder.mkdir(parents=True, exist_ok=True)

# Copying MIDI files from Drive to local storage
# Reading from local disk is much faster than streaming from Drive during training

source_midi_files = []

# Searching recursively to handle subfolders inside the multigenre folder
for pattern in ["*.mid", "*.midi", "*.MID", "*.MIDI"]:
    source_midi_files.extend(list(drive_midi_folder.rglob(pattern)))

if len(source_midi_files) == 0:
    raise ValueError("No MIDI files found. Check the Task 2 Drive folder path.")

# Renaming with a numeric prefix to avoid collisions from files with the same name
for index, file in enumerate(source_midi_files):
    safe_name = file.name.replace("/", "_")
    shutil.copy(file, raw_folder / f"{index:04d}_{safe_name}")

# Collecting and sorting the copied files for consistent ordering
midi_files = []
for pattern in ["*.mid", "*.midi", "*.MID", "*.MIDI"]:
    midi_files.extend(list(raw_folder.glob(pattern)))
midi_files = sorted(midi_files)

# Hyperparameters & Config

lowest_note = 21        # MIDI pitch for the lowest piano key (A0)
note_count = 88         # Total keys on a standard piano
frame_rate = 8          # Time steps per second when converting MIDI → piano roll
sequence_length = 64    # Window length in frames (64 / 8 = 8 seconds per segment)
stride = 32             # Sliding 50% overlap so adjacent windows share context

max_midi_files = 250    # Capping file count to keep processing time manageable
max_segments = 5000     # Total windows across train + test combined
train_ratio = 0.8       # Splitting 80% for training, 20% for testing

batch_size = 32
epochs = 40
learning_rate = 0.001
hidden_size = 128       # LSTM hidden state dimension
latent_size = 32        # Bottleneck dimension — the VAE samples from this space
num_layers = 2          # Stacking two LSTM layers in both encoder and decoder
beta_max = 0.0008       # Maximum KL weight — keeping it small to avoid posterior collapse
z_temperature = 1.5     # Scaling noise during generation to encourage variety

midi_files = midi_files[:max_midi_files]

print("Device:", device)
print("MIDI files copied from Drive:", len(source_midi_files))
print("MIDI files used for training:", len(midi_files))

if len(midi_files) < 20:
    raise ValueError("Use at least 20 MIDI files. Better Task 2 result needs 100 to 300 MIDI files.")


# MIDI → Piano Roll Conversion
# Representing each file as a (T, 88) binary matrix:
# rows = time steps, columns = piano keys, 1 = note active, 0 = note off

def midi_to_roll(path):
    try:
        midi = pretty_midi.PrettyMIDI(str(path))

        # Getting the full 128-pitch piano roll, then slicing to 88 piano keys
        roll = midi.get_piano_roll(fs=frame_rate)
        roll = roll[lowest_note:lowest_note + note_count, :].T

        # Binarizing: ignoring velocity, only caring whether a note is on or off
        roll = (roll > 0).astype(np.float32)
        return roll
    except Exception as error:
        print("Skipped:", path.name, error)
        return None


def make_segments(roll):
    """Slicing a piano roll into overlapping fixed-length windows."""
    segments = []

    if roll is None:
        return segments

    # Skipping files shorter than one full window
    if len(roll) < sequence_length:
        return segments

    for start in range(0, len(roll) - sequence_length + 1, stride):
        segment = roll[start:start + sequence_length]

        # Dropping silent windows — they carry no musical information
        if segment.sum() > 0:
            segments.append(segment)

    return segments


def build_dataset(files, limit):
    """Processing a list of MIDI files into a shuffled array of piano-roll segments."""
    data = []

    for file in tqdm(files, desc="Processing MIDI"):
        roll = midi_to_roll(file)
        segments = make_segments(roll)
        data.extend(segments)

    random.shuffle(data)
    data = data[:limit]  # Trimming to the segment cap

    return np.array(data, dtype=np.float32)


# Train / Test Split (splitting at the file level to avoid data leakage)

random.shuffle(midi_files)

train_file_count = int(len(midi_files) * train_ratio)
train_files = midi_files[:train_file_count]
test_files = midi_files[train_file_count:]

# Handling the edge case where test set ends up empty
if len(test_files) == 0:
    test_files = train_files[-1:]
    train_files = train_files[:-1]

train_limit = int(max_segments * train_ratio)
test_limit = max_segments - train_limit

train_data = build_dataset(train_files, train_limit)
test_data = build_dataset(test_files, test_limit)

if len(train_data) == 0 or len(test_data) == 0:
    raise ValueError("No valid train/test segments found. Use longer MIDI files.")

# Saving processed arrays so re-runs skip expensive MIDI parsing
np.save(processed_folder / "task2_train.npy", train_data)
np.save(processed_folder / "task2_test.npy", test_data)

print("Train MIDI files:", len(train_files))
print("Test MIDI files:", len(test_files))
print("Train shape:", train_data.shape)
print("Test shape:", test_data.shape)


# Visualization Helper

def plot_roll(roll, title, save_path):
    """Displaying a piano roll as a heatmap and saving it to disk."""
    plt.figure(figsize=(12, 5))
    plt.imshow(roll.T, aspect="auto", origin="lower", interpolation="nearest")
    plt.xlabel("Time step")
    plt.ylabel("Piano key")
    plt.title(title)
    plt.colorbar(label="Active note")
    plt.tight_layout()
    plt.savefig(save_path, dpi=200)
    plt.show()


# Doing a quick sanity check before training by visualizing one training segment
plot_roll(train_data[0], "Task 2 Original Training Piano Roll", plot_folder / "task2_original_training_pianoroll.png")


# PyTorch Dataset Wrapper
# Wrapping the numpy array so DataLoader can handle batching and shuffling

class PianoRollDataset(Dataset):
    def __init__(self, data):
        self.data = torch.tensor(data, dtype=torch.float32)

    def __len__(self):
        return len(self.data)

    def __getitem__(self, index):
        return self.data[index]


train_loader = DataLoader(PianoRollDataset(train_data), batch_size=batch_size, shuffle=True)
test_loader = DataLoader(PianoRollDataset(test_data), batch_size=batch_size, shuffle=False)


# LSTM VAE Architecture
#
# Key difference from Task 1 autoencoder:
#   Instead of encoding to a single vector z, we encode to (mu, logvar),
#   then sampling z ~ N(mu, exp(logvar)) using the reparameterization trick.
#   This makes the latent space continuous so we can freely sample from it.
#
#   Encoder: LSTM → hidden state → two linear heads → mu and logvar
#   Sampling: z = mu + std * epsilon  (epsilon ~ N(0,1))
#   Decoder: z → repeated hidden → LSTM → piano roll

class LSTMVAE(nn.Module):
    def __init__(self, input_size, hidden_size, latent_size, num_layers):
        super().__init__()

        # Encoding the input sequence into a fixed-size hidden representation
        self.encoder = nn.LSTM(
            input_size=input_size,
            hidden_size=hidden_size,
            num_layers=num_layers,
            batch_first=True,
            dropout=0.2 if num_layers > 1 else 0.0
        )

        # Projecting the hidden state to the mean of the latent distribution
        self.to_mu = nn.Linear(hidden_size, latent_size)

        # Projecting the hidden state to log-variance (log for numerical stability)
        self.to_logvar = nn.Linear(hidden_size, latent_size)

        # Expanding the sampled z back to hidden size before feeding the decoder
        self.from_latent = nn.Linear(latent_size, hidden_size)

        # Decoding the latent vector back into a piano roll sequence
        self.decoder = nn.LSTM(
            input_size=hidden_size,
            hidden_size=hidden_size,
            num_layers=num_layers,
            batch_first=True,
            dropout=0.2 if num_layers > 1 else 0.0
        )

        # Mapping decoder output back to 88-note piano roll space
        self.to_piano_roll = nn.Linear(hidden_size, input_size)

    def encode(self, x):
        output, hidden_state = self.encoder(x)
        hidden, cell = hidden_state

        # Taking only the top LSTM layer's hidden state as the sequence summary
        last_hidden = hidden[-1]
        mu = self.to_mu(last_hidden)
        logvar = self.to_logvar(last_hidden)
        return mu, logvar

    def reparameterize(self, mu, logvar):
        # Sampling z = mu + std * epsilon — keeping gradients flowing through mu and logvar
        std = torch.exp(0.5 * logvar)
        noise = torch.randn_like(std)
        z = mu + std * noise
        return z

    def decode(self, z, length):
        # Projecting z to hidden size, then repeating it across all time steps
        start = torch.relu(self.from_latent(z))
        repeated = start.unsqueeze(1).repeat(1, length, 1)

        output, hidden_state = self.decoder(repeated)

        # Squashing output to [0,1] to match the binary piano roll targets
        roll = torch.sigmoid(self.to_piano_roll(output))
        return roll

    def forward(self, x):
        mu, logvar = self.encode(x)
        z = self.reparameterize(mu, logvar)
        reconstructed = self.decode(z, x.size(1))
        return reconstructed, mu, logvar, z


# Instantiating model and moving to GPU if available
model = LSTMVAE(
    input_size=note_count,
    hidden_size=hidden_size,
    latent_size=latent_size,
    num_layers=num_layers
).to(device)

optimizer = optim.Adam(model.parameters(), lr=learning_rate)

print(model)


# VAE Loss Function
# Combining two terms:
#   1. Reconstruction loss — measuring how well the decoder reproduces the input
#   2. KL divergence — pushing the latent distribution toward N(0,1)
#      so the latent space stays continuous and we can sample from it freely
# Beta controls the balance. Starting small and annealing up avoids
# the model collapsing to ignoring the latent vector entirely (posterior collapse).

def vae_loss(reconstructed, original, mu, logvar, beta):
    reconstruction_loss = torch.mean((reconstructed - original) ** 2)

    # Summing across latent dims, then averaging across the batch
    kl_loss = -0.5 * torch.mean(torch.sum(1 + logvar - mu.pow(2) - logvar.exp(), dim=1))

    total_loss = reconstruction_loss + beta * kl_loss
    return total_loss, reconstruction_loss, kl_loss


# Tracking all three loss components separately for plotting later
train_total_losses = []
train_reconstruction_losses = []
train_kl_losses = []

test_total_losses = []
test_reconstruction_losses = []
test_kl_losses = []

# Training Loop

for epoch in range(1, epochs + 1):
    model.train()

    # Linearly annealing beta from 0 → beta_max over the first 12 epochs
    # This lets the model learn to reconstruct well before the KL term kicks in
    beta = beta_max * min(1.0, epoch / 12)

    train_total_sum = 0
    train_recon_sum = 0
    train_kl_sum = 0

    for batch in train_loader:
        batch = batch.to(device)

        reconstructed, mu, logvar, z = model(batch)
        loss, recon_loss, kl_loss = vae_loss(reconstructed, batch, mu, logvar, beta)

        optimizer.zero_grad()
        loss.backward()

        # Clipping gradients to prevent exploding gradient issues common in RNNs
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()

        # Accumulating weighted by batch size for a correct per-sample average later
        train_total_sum += loss.item() * batch.size(0)
        train_recon_sum += recon_loss.item() * batch.size(0)
        train_kl_sum += kl_loss.item() * batch.size(0)

    # Normalizing by dataset size to get the average loss per sample
    train_total = train_total_sum / len(train_loader.dataset)
    train_recon = train_recon_sum / len(train_loader.dataset)
    train_kl = train_kl_sum / len(train_loader.dataset)

    # --- Evaluating on test set (no gradient computation needed) ---
    model.eval()
    test_total_sum = 0
    test_recon_sum = 0
    test_kl_sum = 0

    with torch.no_grad():
        for batch in test_loader:
            batch = batch.to(device)
            reconstructed, mu, logvar, z = model(batch)
            loss, recon_loss, kl_loss = vae_loss(reconstructed, batch, mu, logvar, beta)

            test_total_sum += loss.item() * batch.size(0)
            test_recon_sum += recon_loss.item() * batch.size(0)
            test_kl_sum += kl_loss.item() * batch.size(0)

    test_total = test_total_sum / len(test_loader.dataset)
    test_recon = test_recon_sum / len(test_loader.dataset)
    test_kl = test_kl_sum / len(test_loader.dataset)

    train_total_losses.append(train_total)
    train_reconstruction_losses.append(train_recon)
    train_kl_losses.append(train_kl)

    test_total_losses.append(test_total)
    test_reconstruction_losses.append(test_recon)
    test_kl_losses.append(test_kl)

    # Printing at epoch 1, every 5 epochs, and at the end
    if epoch == 1 or epoch % 5 == 0 or epoch == epochs:
        print(
            f"Epoch {epoch:02d}/{epochs} | "
            f"Beta: {beta:.6f} | "
            f"Train total: {train_total:.6f} | "
            f"Train recon: {train_recon:.6f} | "
            f"Train KL: {train_kl:.6f} | "
            f"Test total: {test_total:.6f}"
        )


# Plotting three separate loss curves to understand what each term is doing
plt.figure(figsize=(9, 5))
plt.plot(train_total_losses, label="Train total loss")
plt.plot(test_total_losses, label="Test total loss")
plt.xlabel("Epoch")
plt.ylabel("VAE loss")
plt.title("Task 2 VAE Total Loss")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig(plot_folder / "task2_vae_total_loss_curve.png", dpi=200)
plt.show()

plt.figure(figsize=(9, 5))
plt.plot(train_reconstruction_losses, label="Train reconstruction loss")
plt.plot(test_reconstruction_losses, label="Test reconstruction loss")
plt.xlabel("Epoch")
plt.ylabel("MSE reconstruction loss")
plt.title("Task 2 VAE Reconstruction Loss")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig(plot_folder / "task2_vae_reconstruction_loss_curve.png", dpi=200)
plt.show()

# KL loss rising too fast = posterior collapse; staying near zero = VAE ignoring prior
plt.figure(figsize=(9, 5))
plt.plot(train_kl_losses, label="Train KL loss")
plt.plot(test_kl_losses, label="Test KL loss")
plt.xlabel("Epoch")
plt.ylabel("KL divergence")
plt.title("Task 2 VAE KL-Divergence Loss")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig(plot_folder / "task2_vae_kl_loss_curve.png", dpi=200)
plt.show()

# Reconstruction Quality Check
# Running one test batch and visually inspecting original vs reconstructed

model.eval()

with torch.no_grad():
    batch = next(iter(test_loader)).to(device)
    reconstructed, mu, logvar, z = model(batch)

original_sample = batch[0].cpu().numpy()
reconstructed_sample = reconstructed[0].cpu().numpy()
difference = np.abs(original_sample - reconstructed_sample)  # pixel-wise absolute error

plot_roll(original_sample, "Task 2 Original Test Segment", plot_folder / "task2_original_test_segment.png")
plot_roll(reconstructed_sample, "Task 2 Reconstructed Test Segment", plot_folder / "task2_reconstructed_test_segment.png")
plot_roll(difference, "Task 2 Reconstruction Difference", plot_folder / "task2_reconstruction_difference.png")


# Piano Roll → MIDI Conversion
# Converting a binary (0/1) piano roll matrix back into a playable MIDI file

def piano_roll_to_midi_from_binary(binary, save_path, velocity=90):
    midi = pretty_midi.PrettyMIDI(initial_tempo=120)
    piano = pretty_midi.Instrument(program=0, name="Generated Piano")  # program=0 = acoustic grand piano

    # Padding with silence on both ends so diff catches edge-of-array note transitions
    padded = np.pad(binary, ((1, 1), (0, 0)), mode="constant")
    changes = np.diff(padded, axis=0)

    for note_index in range(note_count):
        pitch = note_index + lowest_note
        starts = np.where(changes[:, note_index] == 1)[0]
        ends = np.where(changes[:, note_index] == -1)[0]

        for start, end in zip(starts, ends):
            start_time = start / frame_rate
            end_time = max(end / frame_rate, start_time + 0.05)  # enforcing minimum 50ms note length

            note = pretty_midi.Note(
                velocity=int(velocity),
                pitch=int(pitch),
                start=float(start_time),
                end=float(end_time)
            )
            piano.notes.append(note)

    midi.instruments.append(piano)
    midi.write(str(save_path))

    return len(piano.notes)


# Adaptive Binarization
# Instead of a fixed 0.5 threshold, adjusting the cutoff so the output
# note density roughly matches the source piece (avoiding too sparse/dense output)

def sample_binary_from_probability(roll, target_density):
    # Clamping to a musically sensible range (1.5% – 9% of cells active)
    target_density = float(np.clip(target_density, 0.015, 0.09))

    # Normalizing the roll to [0, 1] before thresholding
    adjusted = roll.copy()
    adjusted = adjusted - adjusted.min()

    if adjusted.max() > 0:
        adjusted = adjusted / adjusted.max()

    # Applying a random temperature to vary sharpness of note decisions
    temperature = np.random.uniform(0.7, 1.6)
    adjusted = adjusted ** temperature

    # Setting threshold so the top `target_density` fraction of cells are active
    threshold = np.quantile(adjusted.reshape(-1), 1.0 - target_density)
    binary = (adjusted > threshold).astype(np.int32)

    # Randomly dropping some active notes to add more natural variation
    keep_probability = np.random.uniform(0.80, 0.95)
    random_keep = np.random.rand(*binary.shape) < keep_probability
    binary = (binary * random_keep).astype(np.int32)

    return binary, threshold


# Diverse VAE Sample Generation
# Using three different sampling strategies to maximize variety:
#   "random"         — sampling z directly from N(0, scale)
#   "encoded_noise"  — encoding a real piece then adding noise to its mu
#   "interpolation"  — linearly blending the mus of two real pieces
# Rejecting duplicate outputs using MD5 fingerprinting.

def generate_more_diverse_vae_samples(sample_count=8):
    model.eval()

    pairs = []
    info = []
    fingerprints = set()
    average_density = float(train_data.mean())
    attempts = 0

    while len(pairs) < sample_count and attempts < 250:
        attempts += 1

        with torch.no_grad():
            mode = np.random.choice(["random", "encoded_noise", "interpolation"])

            if mode == "random":
                # Sampling purely from the prior — most creative but least structured
                z = torch.randn(1, latent_size).to(device) * np.random.uniform(1.0, 3.0)

            elif mode == "encoded_noise":
                # Encoding a real training piece and perturbing it in latent space
                idx = np.random.randint(0, len(train_data))
                real_roll = torch.tensor(train_data[idx], dtype=torch.float32).unsqueeze(0).to(device)
                mu, logvar = model.encode(real_roll)
                z = mu + torch.randn_like(mu) * np.random.uniform(1.0, 3.0)

            else:
                # Interpolating between two real pieces in latent space
                idx1 = np.random.randint(0, len(train_data))
                idx2 = np.random.randint(0, len(train_data))

                x1 = torch.tensor(train_data[idx1], dtype=torch.float32).unsqueeze(0).to(device)
                x2 = torch.tensor(train_data[idx2], dtype=torch.float32).unsqueeze(0).to(device)

                mu1, logvar1 = model.encode(x1)
                mu2, logvar2 = model.encode(x2)

                # Blending with a random alpha to explore the space between them
                alpha = np.random.uniform(0.15, 0.85)
                z = (1 - alpha) * mu1 + alpha * mu2
                z = z + torch.randn_like(z) * np.random.uniform(0.8, 2.2)

            roll = model.decode(z, sequence_length).cpu().numpy()[0]

        target_density = average_density * np.random.uniform(0.6, 2.2)
        binary, threshold = sample_binary_from_probability(roll, target_density)

        active_cells = int(binary.sum())
        fingerprint = hashlib.md5(binary.tobytes()).hexdigest()

        # Keeping only non-empty, non-duplicate outputs
        if active_cells > 0 and fingerprint not in fingerprints:
            fingerprints.add(fingerprint)
            pairs.append((roll, binary))
            info.append({
                "sample_mode": mode,
                "active_cells": active_cells,
                "threshold": float(threshold),
                "target_density": float(target_density)
            })

    if len(pairs) < sample_count:
        raise ValueError("Could not generate enough diverse samples. Try running this cell again.")

    return pairs, info


# Generating 8 diverse MIDI samples and exporting them
generated_pairs, generation_details = generate_more_diverse_vae_samples(8)

generated_samples = []
generated_binary = []
generated_info = []

for i, pair in enumerate(generated_pairs, start=1):
    roll, binary = pair

    save_path = midi_output_folder / f"task2_vae_generated_{i}.mid"
    note_total = piano_roll_to_midi_from_binary(binary, save_path)

    generated_samples.append(roll)
    generated_binary.append(binary.astype(np.float32))

    plot_roll(binary, f"Task 2 Diverse VAE Generated Sample {i}", plot_folder / f"task2_vae_generated_sample_{i}.png")

    generated_info.append({
        "sample": f"task2_vae_generated_{i}.mid",
        "notes": note_total,
        "sample_mode": generation_details[i - 1]["sample_mode"],
        "active_cells": generation_details[i - 1]["active_cells"],
        "threshold": generation_details[i - 1]["threshold"],
        "target_density": generation_details[i - 1]["target_density"]
    })

    print(
        f"Generated {save_path.name} | "
        f"notes: {note_total} | "
        f"mode: {generation_details[i - 1]['sample_mode']}"
    )


# Latent Space Interpolation
# Picking two training pieces, encoding them to mu1 and mu2,
# then smoothly stepping from one to the other with alpha = 0, 0.25, 0.5, 0.75, 1.0.
# Useful for listening to how the VAE transitions between two musical styles.

def latent_interpolation(count):
    model.eval()
    items = []
    average_density = float(train_data.mean())

    with torch.no_grad():
        idx1 = np.random.randint(0, len(train_data))
        idx2 = np.random.randint(0, len(train_data))

        x1 = torch.tensor(train_data[idx1], dtype=torch.float32).unsqueeze(0).to(device)
        x2 = torch.tensor(train_data[idx2], dtype=torch.float32).unsqueeze(0).to(device)

        mu1, logvar1 = model.encode(x1)
        mu2, logvar2 = model.encode(x2)

        for i in range(count):
            # Evenly spacing alpha from 0 (= first piece) to 1 (= second piece)
            alpha = i / (count - 1)
            z = (1 - alpha) * mu1 + alpha * mu2

            # Adding a small amount of noise to avoid identical adjacent outputs
            z = z + torch.randn_like(z) * 0.4

            roll = model.decode(z, sequence_length).cpu().numpy()[0]
            target_density = average_density * np.random.uniform(0.8, 1.7)
            binary, threshold = sample_binary_from_probability(roll, target_density)
            items.append((alpha, roll, binary, threshold, target_density))

    return items


interpolation_samples = latent_interpolation(5)
interpolation_info = []

for i, item in enumerate(interpolation_samples, start=1):
    alpha, roll, binary, threshold, target_density = item
    save_path = midi_output_folder / f"task2_latent_interpolation_{i}.mid"
    note_total = piano_roll_to_midi_from_binary(binary, save_path)

    plot_roll(binary, f"Task 2 Latent Interpolation {i}", plot_folder / f"task2_latent_interpolation_{i}.png")

    interpolation_info.append({
        "sample": f"task2_latent_interpolation_{i}.mid",
        "alpha": alpha,
        "notes": note_total,
        "threshold": threshold,
        "target_density": target_density
    })

    print(f"Interpolation {i} | alpha: {alpha:.2f} | notes: {note_total}")


# Baseline Models (used for comparison with the VAE in the evaluation table)

def random_baseline(sample_count):
    """Generating purely random notes at the same average density as the training data."""
    density = float(train_data.mean())
    samples = []

    for i in range(sample_count):
        sample = (np.random.rand(sequence_length, note_count) < density).astype(np.float32)
        samples.append(sample)

    return samples


def markov_probabilities(data):
    """
    Computing per-note first-order Markov transition probabilities from training data.
    Returning shape (88, 2, 2): for each note, P(next_state | prev_state).
    """
    probabilities = np.zeros((note_count, 2, 2), dtype=np.float32)

    for note in range(note_count):
        previous_values = data[:, :-1, note].reshape(-1).astype(int)
        next_values = data[:, 1:, note].reshape(-1).astype(int)

        for previous_state in [0, 1]:
            mask = previous_values == previous_state
            count = mask.sum()

            if count == 0:
                # Defaulting to always-off if this transition was never seen in training
                probabilities[note, previous_state, 0] = 1.0
                probabilities[note, previous_state, 1] = 0.0
            else:
                p_on = next_values[mask].mean()
                probabilities[note, previous_state, 1] = p_on
                probabilities[note, previous_state, 0] = 1.0 - p_on

    return probabilities


def markov_baseline(sample_count):
    """Generating samples by rolling forward per-note Markov chains learned from training data."""
    probabilities = markov_probabilities(train_data)
    start_probability = train_data[:, 0, :].mean(axis=0)  # marginal probability of each note at t=0
    samples = []

    for i in range(sample_count):
        sample = np.zeros((sequence_length, note_count), dtype=np.float32)

        # Sampling the first frame from training statistics
        sample[0] = (np.random.rand(note_count) < start_probability).astype(np.float32)

        # Rolling each note forward independently through its learned transitions
        for t in range(1, sequence_length):
            for note in range(note_count):
                previous_state = int(sample[t - 1, note])
                p_on = probabilities[note, previous_state, 1]
                sample[t, note] = 1.0 if np.random.rand() < p_on else 0.0

        samples.append(sample)

    return samples


# Generating and exporting baseline samples
random_samples = random_baseline(8)
markov_samples = markov_baseline(8)

for i, sample in enumerate(random_samples, start=1):
    piano_roll_to_midi_from_binary(sample.astype(np.int32), midi_output_folder / f"task2_baseline_random_{i}.mid")

for i, sample in enumerate(markov_samples, start=1):
    piano_roll_to_midi_from_binary(sample.astype(np.int32), midi_output_folder / f"task2_baseline_markov_{i}.mid")


# Evaluation Metrics
# Comparing all three models on four musical properties

def pitch_histogram_from_binary(binary_samples):
    """
    Computing a normalized 12-bin pitch class histogram (collapsing octaves).
    Measuring which note names (C, C#, D, ...) appear most across the samples.
    """
    histogram = np.zeros(12, dtype=np.float64)

    for binary in binary_samples:
        active = binary.sum(axis=0)

        for note_index, count in enumerate(active):
            pitch = note_index + lowest_note
            histogram[pitch % 12] += count

    if histogram.sum() == 0:
        return histogram

    return histogram / histogram.sum()


def pitch_distance(reference_binary, generated_binary):
    """Computing L1 distance between pitch class histograms. Lower = closer to real music."""
    p = pitch_histogram_from_binary(reference_binary)
    q = pitch_histogram_from_binary(generated_binary)
    return float(np.sum(np.abs(p - q)))


def durations_from_binary(binary):
    """Extracting the duration in frames of every note event in a binary piano roll."""
    binary = binary.astype(np.int32)
    durations = []

    padded = np.pad(binary, ((1, 1), (0, 0)), mode="constant")
    changes = np.diff(padded, axis=0)

    for note in range(binary.shape[1]):
        starts = np.where(changes[:, note] == 1)[0]
        ends = np.where(changes[:, note] == -1)[0]

        for start, end in zip(starts, ends):
            durations.append(end - start)

    return durations


def rhythm_diversity(binary_samples):
    """
    Computing the ratio of unique note durations to total note events.
    Higher = more varied rhythm (0 = all same length, 1 = every note unique).
    """
    all_durations = []

    for binary in binary_samples:
        all_durations.extend(durations_from_binary(binary))

    if len(all_durations) == 0:
        return 0.0

    return float(len(set(all_durations)) / len(all_durations))


def repetition_ratio(binary_samples, pattern_length=8):
    """
    Computing the fraction of 8-frame windows that are exact repeats within the same sample.
    Higher = more repetitive / less creative output.
    """
    repeated = 0
    total = 0

    for binary in binary_samples:
        binary = binary.astype(np.int32)
        patterns = []

        for start in range(0, len(binary) - pattern_length + 1, pattern_length):
            pattern = tuple(binary[start:start + pattern_length].reshape(-1).tolist())
            patterns.append(pattern)

        total += len(patterns)
        repeated += len(patterns) - len(set(patterns))

    if total == 0:
        return 0.0

    return float(repeated / total)


def note_density(binary_samples):
    """Computing the average fraction of active notes across all samples and time steps."""
    return float(np.mean([binary.mean() for binary in binary_samples]))


# Binarizing all sample sets at 0.5 for a fair apples-to-apples comparison
reference_binary = [(roll > 0.5).astype(np.float32) for roll in test_data[:min(200, len(test_data))]]
random_binary = [roll.astype(np.float32) for roll in random_samples]
markov_binary = [roll.astype(np.float32) for roll in markov_samples]

# Building the comparison table
task2_results = [
    {
        "Model": "Random Note Generator",
        "Pitch Histogram Distance": pitch_distance(reference_binary, random_binary),
        "Rhythm Diversity": rhythm_diversity(random_binary),
        "Repetition Ratio": repetition_ratio(random_binary),
        "Average Note Density": note_density(random_binary)
    },
    {
        "Model": "Markov Chain Baseline",
        "Pitch Histogram Distance": pitch_distance(reference_binary, markov_binary),
        "Rhythm Diversity": rhythm_diversity(markov_binary),
        "Repetition Ratio": repetition_ratio(markov_binary),
        "Average Note Density": note_density(markov_binary)
    },
    {
        "Model": "Task 2 VAE",
        "Pitch Histogram Distance": pitch_distance(reference_binary, generated_binary),
        "Rhythm Diversity": rhythm_diversity(generated_binary),
        "Repetition Ratio": repetition_ratio(generated_binary),
        "Average Note Density": note_density(generated_binary)
    }
]

task2_results_df = pd.DataFrame(task2_results)
task2_results_df.to_csv(result_folder / "task2_metrics.csv", index=False)

generated_df = pd.DataFrame(generated_info)
generated_df.to_csv(result_folder / "task2_generated_samples_info.csv", index=False)

interpolation_df = pd.DataFrame(interpolation_info)
interpolation_df.to_csv(result_folder / "task2_latent_interpolation_info.csv", index=False)

display(task2_results_df)
display(generated_df)
display(interpolation_df)

# Plotting a bar chart for each metric for a quick visual comparison
for column in task2_results_df.columns:
    if column == "Model":
        continue

    plt.figure(figsize=(8, 4))
    plt.bar(task2_results_df["Model"], task2_results_df[column])
    plt.xticks(rotation=20, ha="right")
    plt.ylabel(column)
    plt.title(f"Task 2 {column}")
    plt.tight_layout()
    plt.savefig(plot_folder / f"task2_{column.lower().replace(' ', '_')}.png", dpi=200)
    plt.show()


# Task 1 vs Task 2 Comparison
# Pulling the Task 1 metrics CSV from the saved ZIP and merging it with
# the Task 2 results for a side-by-side comparison table

task1_comparison_df = None

if task1_zip_path.exists():
    with zipfile.ZipFile(task1_zip_path, "r") as z:
        metric_files = [name for name in z.namelist() if name.endswith("task1_metrics.csv")]

        if len(metric_files) > 0:
            extract_folder = Path("/content/task1_compare")

            if extract_folder.exists():
                shutil.rmtree(extract_folder)

            extract_folder.mkdir(parents=True, exist_ok=True)

            # Extracting only the metrics CSV — no need to unzip the whole archive
            z.extract(metric_files[0], extract_folder)
            task1_metric_path = extract_folder / metric_files[0]
            task1_df = pd.read_csv(task1_metric_path)

            # Finding the LSTM Autoencoder row from Task 1
            task1_model_rows = task1_df[task1_df["Model"].astype(str).str.contains("Autoencoder", case=False, na=False)]

            if len(task1_model_rows) > 0:
                task1_row = task1_model_rows.iloc[0].to_dict()
                task2_row = task2_results_df[task2_results_df["Model"] == "Task 2 VAE"].iloc[0].to_dict()

                comparison_rows = [
                    {"Task": "Task 1 LSTM Autoencoder", **task1_row},
                    {"Task": "Task 2 VAE", **task2_row}
                ]

                task1_comparison_df = pd.DataFrame(comparison_rows)
                task1_comparison_df.to_csv(result_folder / "task1_vs_task2_metrics_comparison.csv", index=False)
                display(task1_comparison_df)

if task1_comparison_df is None:
    print("Task 1 vs Task 2 automatic comparison was not created. Compare both CSV files manually in the report.")


# Saving Model Checkpoint
# Bundling weights, all loss histories, and hyperparameters into a single file

torch.save(
    {
        "model": model.state_dict(),
        "train_total_losses": train_total_losses,
        "train_reconstruction_losses": train_reconstruction_losses,
        "train_kl_losses": train_kl_losses,
        "test_total_losses": test_total_losses,
        "test_reconstruction_losses": test_reconstruction_losses,
        "test_kl_losses": test_kl_losses,
        "settings": {
            "lowest_note": lowest_note,
            "note_count": note_count,
            "frame_rate": frame_rate,
            "sequence_length": sequence_length,
            "hidden_size": hidden_size,
            "latent_size": latent_size,
            "num_layers": num_layers,
            "beta_max": beta_max
        }
    },
    model_folder / "task2_lstm_vae.pth"
)

# Summary, README, and ZIP Export

summary = pd.DataFrame([{
    "Task": "Task 2 VAE Music Generator",
    "Dataset": "Lakh MIDI Clean subset",
    "MIDI files used": len(midi_files),
    "Train MIDI files": len(train_files),
    "Test MIDI files": len(test_files),
    "Train segments": len(train_data),
    "Test segments": len(test_data),
    "Final train total loss": train_total_losses[-1],
    "Final train reconstruction loss": train_reconstruction_losses[-1],
    "Final train KL loss": train_kl_losses[-1],
    "Final test total loss": test_total_losses[-1],
    "Generated MIDI samples": 8,
    "Latent interpolation samples": 5
}])

summary.to_csv(result_folder / "task2_summary.csv", index=False)
display(summary)

requirements_text = """pretty_midi
mido
tqdm
numpy
pandas
matplotlib
torch
"""

readme_text = """# CSE425 Task 2: VAE Music Generator

This repository contains Task 2 of the Neural Networks project. The goal is to generate diverse MIDI music samples using an unsupervised Variational Autoencoder.

## Dataset

The dataset used is a subset of the Lakh MIDI Clean dataset. The full dataset is not uploaded because of size. MIDI files were stored in Google Drive and loaded into Google Colab.

## Method

MIDI files were converted into 88-note binary piano-roll representation. The piano-rolls were divided into fixed-length windows. The LSTM encoder produced mean and log-variance vectors. A latent vector was sampled using the reparameterization trick, and the LSTM decoder reconstructed the piano-roll sequence.

The loss function combines reconstruction loss and KL-divergence.

## Outputs

The project includes:
- LSTM VAE implementation
- Total loss curve
- Reconstruction loss curve
- KL-divergence loss curve
- 8 generated MIDI samples
- 5 latent interpolation MIDI samples
- Random Note Generator baseline
- Markov Chain baseline
- Evaluation metrics table
- Task 1 vs Task 2 comparison if Task 1 output ZIP is available

## How to Run

1. Upload MIDI files to Google Drive:
   MyDrive/CSE425_Project/task2_midi/multigenre/

2. Open the notebook in Google Colab.

3. Run all cells.

4. Outputs will be generated inside:
   CSE425_Task2_VAE_Music_Generator/outputs/
"""

(project_folder / "requirements.txt").write_text(requirements_text)
(project_folder / "README.md").write_text(readme_text)

# Packaging everything into a ZIP and copying to Drive for easy access
zip_path = "/content/CSE425_Task2_VAE_Music_Generator.zip"

if Path(zip_path).exists():
    Path(zip_path).unlink()

shutil.make_archive("/content/CSE425_Task2_VAE_Music_Generator", "zip", project_folder)

drive_output_folder = Path("/content/drive/MyDrive/CSE425_Project/task2_outputs")
drive_output_folder.mkdir(parents=True, exist_ok=True)

shutil.copy(zip_path, drive_output_folder / "CSE425_Task2_VAE_Music_Generator.zip")

print("Task 2 finished.")
print("Generated MIDI:", midi_output_folder)
print("Plots:", plot_folder)
print("Results:", result_folder)
print("ZIP file in Colab:", zip_path)
print("ZIP file copied to Drive:", drive_output_folder / "CSE425_Task2_VAE_Music_Generator.zip")
```
